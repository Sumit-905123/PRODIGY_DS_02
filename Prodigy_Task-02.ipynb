```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = pd.read_csv('gender_submission.csv')
titanic = pd.read_csv('test.csv')
titanic = pd.read_csv('train.csv')

# Display the first few rows of the dataset
titanic.head()
# Check for missing values
titanic.isnull().sum()
# Check for missing values
titanic.isnull().sum()
# Check data types
titanic.dtypes
# Convert 'Survived', 'Pclass', and 'Embarked' to category
titanic['Survived'] = titanic['Survived'].astype('category')
titanic['Pclass'] = titanic['Pclass'].astype('category')
titanic['Embarked'] = titanic['Embarked'].astype('category')
# Summary statistics
titanic.describe(include='all')
# Survived
sns.countplot(x='Survived', data=titanic)
plt.title('Survival Count')
plt.show()


# Sex
sns.countplot(x='Sex', data=titanic)
plt.title('Gender Count')
plt.show()

# Embarked
sns.countplot(x='Embarked', data=titanic)
plt.title('Port of Embarkation Count')
plt.show()
# Age
sns.histplot(titanic['Age'], kde=True)
plt.title('Age Distribution')
plt.show()

# Fare
sns.histplot(titanic['Fare'], kde=True)
plt.title('Fare Distribution')
plt.show()

# Survival by Sex
sns.countplot(x='Sex', hue='Survived', data=titanic)
plt.title('Survival by Gender')
plt.show()

# Survival by Age
sns.histplot(data=titanic, x='Age', hue='Survived', multiple='stack')
plt.title('Survival by Age')
plt.show()

# Survival by Embarked
sns.countplot(x='Embarked', hue='Survived', data=titanic)
plt.title('Survival by Port of Embarkation')
plt.show()

# Fare vs Survived
sns.boxplot(x='Survived', y='Fare', data=titanic)
plt.title('Fare vs Survival')
plt.show()
# Pairplot for numerical variables
sns.pairplot(titanic[['Age', 'Fare', 'Pclass', 'Survived']], hue='Survived')
plt.show()

# Heatmap for correlation matrix
corr_matrix = titanic.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Drop non-numeric columns
numeric_columns = titanic.select_dtypes(include=[np.number]).columns
titanic_numeric = titanic[numeric_columns]

# Calculate correlation matrix
corr_matrix = titanic_numeric.corr()

# Plot heatmap for correlation matrix
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
